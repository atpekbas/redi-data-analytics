Create a scatter plot on a world map using the Latitude and Longitude columns to position the markers. The size and color of the markers should represent the Magnitude values. 

// hint: 

Use plotly

The color_continuous_scale parameter determines the color scale used for the Magnitude values. The projection parameter sets the map projection type, and the update_geos() function adds country borders to the map.



```
fig = px.scatter_geo(df,
                     lat='Latitude',
                     lon='Longitude',
                     size='Magnitude',
                     color='Magnitude',
                     color_continuous_scale=px.colors.sequential.Plasma,
                     projection='natural earth',
                     labels={'Latitude': 'Latitude', 'Longitude': 'Longitude', 'Magnitude': 'Magnitude'},
                     title='Earthquakes')

fig.update_geos(showcountries=True, countrywidth=0.5)
```
