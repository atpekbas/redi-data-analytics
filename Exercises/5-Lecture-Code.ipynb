### 3. Analysing the Data

After you are confirmed that the data was loaded correctly, you always want to start with getting to know the dataset.
```python
df_raw.info()
```

You can use the following `show_details(dataframe)` method to show details about the dataframe.<br>

```python
def isNull(x):
       return x.isna().sum()
        
def show_details(dataframe):
    return dataframe.agg(['dtype', 'count', 'nunique', isNull])
    
show_details(df_raw)
```

### 4. Cleaning the Data

After identifying the general characteristics of our dataset, we can clean it up. Below, you can see a couple of queries that you can perform to make your data cleaner. Clean data is the fundation of a good data analysis!
Cleaning the dataset can include, but is not limited to:
* dealing with missing values
* formating date columns
* removing unnecessary columns
* adding calculated columns
* changing datatypes

After we are satisfied with the cleanliness of our dataset, we should always save it under a new name. This allows you to look back at the original dataset easily and have a clear separation between the cleaned and raw dataset. You do not have to run the cleaning code again, and can simply work with the already cleaned dataset next time you open up the code.

```python
# Remove rows that have null values in one of the defined columns
df = df_raw.dropna(subset=['date_added', 'rating', 'duration', 'country'])
# Remove the defined columns
df = df.drop(['director', 'cast', 'description'], axis=1)
```

```python
# Parse date_added to datetimes
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), format='%B %d, %Y')
```

```python
# Cast durations to int
isMovie = df['type'] == 'Movie'
df.loc[isMovie, 'duration'] = df.loc[isMovie, 'duration'].str.split(' ').str[0]
df.loc[~isMovie, 'duration'] = df.loc[~isMovie, 'duration'].str.split(' ').str[0]
df.duration = df.duration.astype(int)
```

```python
# Create new column with number of countries
df['no_countries'] = df['country'].str.split(',').str.len()
```

```python
df['main_country'] = df['country'].str.split(',').str[0]
```

```python
df.agg(['dtype', 'count', 'nunique', isNull])
```

```python
df.to_csv('data/netflix_titles_cleaned.csv', index=False)
```

### 6. Visualisation

The visualisation part of data analysis is arguably one of the most important ones. This is what managers, CEOs and stakeholders use to make decisions. You should always structure your visuals so that they are clear and understandable. Preferably have a meaningful title. The colors should be meaningful and consistent. We have given you multiple links to websites that can help you figure out what kind of visuals you need to use for different analysis; please use these.

Below you see a couple of very basic visuals that can be done. Be creative!

```python
# plotting the distribution of movies and TV shows
df['type'].value_counts().plot.bar()
```

```python
# plotting the average duration of movies and TV shows
df.groupby('type')['duration'] \
    .mean(numeric_only=True) \
    .plot.bar()
```

```python
# plotting the distribution of titles by country of release
df.groupby('main_country')['title'] \
    .size() \
    .nlargest(10) \
    .sort_values(ascending=True) \
    .plot.barh()
```

```python
# plotting the change in duration over the years for movies and TV shows
isMovie = df['type'] == 'Movie'
df[isMovie].groupby('release_year')['duration'].mean().reset_index().sort_values('release_year').plot(x='release_year', y='duration')
df[~isMovie].groupby('release_year')['duration'].mean().reset_index().sort_values('release_year').plot(x='release_year', y='duration')
```

```python
# plotting the change in the length of the titles over time
def meanLength(x):
    return x.str.len().mean()

ax = df[isMovie].groupby('release_year')[['title']] \
    .agg(meanLength) \
    .reset_index() \
    .plot(x='release_year', y='title')

ax.set_ylabel('mean title length')
```
