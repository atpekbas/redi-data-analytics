```python
import pandas as pd
df = pd.read_csv("Gayathri Venkatraman - netflix_titles.csv")
df
```

```python
df.info()
```

```python
df.date_added = df.date_added.str.lstrip()
```

```python
df
```

```python
df.date_added =  pd.to_datetime(df.date_added,format = '%B %d, %Y')
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
movie_type = df.loc[df['type']=='Movie']
```

```python
movie_type
```

```python
movie_type.duration = movie_type.duration.str.replace('min','')
```

```python
movie_type['duration'] = pd.to_datetime(movie_type.duration, unit='m').dt.strftime('%H:%M')
```

```python
movie_type['duration'] = pd.to_timedelta(movie_type['duration']+':00')
```

```python
movie_type.info()
```

```python
grouped = movie_type.groupby(["release_year","listed_in"])
```

```python
grouped.get_group((2018,"Dramas, International Movies, Music & Musicals"))
```

```python
grouped.size()
```

```python
grouped.duration.agg('min')
```

```python
grouped.duration.agg('max')
```
