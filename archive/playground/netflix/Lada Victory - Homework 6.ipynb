```python
df["date_added"] = pd.to_datetime(df['date_added'])

df['month_added']=df['date_added'].dt.month
df['month_name_added']=df['date_added'].dt.month_name()
df['year_added'] = df['date_added'].dt.year

df.head(3)
```

```python
df.isnull().sum()
```

```python
df=df.fillna('Not specified')
df.head()
```

###  Split columns.

```python
df.groupby(["type", "country"])[["title"]].count()
```

```python
director_name=df['director'].str.split(',',expand=True).stack()#str.split() with expand=True option results in a data frame
director_name=director_name.to_frame()
director_name.columns=['Director']
directors=director_name.groupby(['Director']).size().reset_index(name='Total Content')
directors=directors[directors.Director !='Not specified']
directors=directors.sort_values(by=['Total Content'],ascending=False)
directors.head()
```

```python
cast_name=pd.DataFrame()
cast_name=df['cast'].str.split(',',expand=True).stack()
cast_name=cast_name.to_frame()
cast_name.columns=['Actor']
actors=cast_name.groupby(['Actor']).size().reset_index(name='Total Content')
actors=actors[actors.Actor !='Not specified']
actors=actors.sort_values(by=['Total Content'],ascending=False)
actors.head()
```

```python
genre_name=pd.DataFrame()
genre_name=df['listed_in'].str.split(',',expand=True).stack()
genre_name=genre_name.to_frame()
genre_name.columns=['Genre']
genre=genre_name.groupby(['Genre']).size().reset_index(name='Total Content')
genre=genre[genre.Genre !='Not specified']
genre=genre.sort_values(by=['Total Content'],ascending=False)
genre.head()
```

```python
#compare the number of TV shows and Movies released over the years after 2020.

release_year=df.groupby(['release_year','type']).size().reset_index(name='Total Content')
release_year=release_year[release_year['release_year'] >= 2020]
release_year=release_year.sort_values(by=['Total Content'],ascending=True)
release_year.head()
```

```python
ratings = df.groupby(['rating']).size().reset_index(name='Total Content')
ratings = ratings[~ratings.rating.str.contains('min')]
ratings = ratings.sort_values(by=['Total Content'],ascending=False)
ratings
```
