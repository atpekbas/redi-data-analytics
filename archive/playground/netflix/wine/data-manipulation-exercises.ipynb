#### Since this variety has been rated so poorly, we come back to the original dataset and reduce the price for all the wines from this variety by 20%...

```python
#wine = pd.read_csv('data/wine.csv')

mask1 = (wine["variety"] == "Red Blend")
mask2 = (wine["points"] <= 85)
mask3 = (wine["country"] == "Italy")


wine.loc[mask1 & mask2 & mask3]
```

```python
### Making 3 masks for each of our conditions
mask1 = (wine["variety"] == "Red Blend")
mask2 = (wine["points"] <= 85)
mask3 = (wine["country"] == "Italy")

### Apllying the masks with loc so we can identify the right rows that we want and then access the value in the column "price" for these rows easily
### Notice how we used wine.price * 0.8 and that's because it will only use the price of the row we have already accessed so we don't need to specify more than wine.price
wine.loc[mask1 & mask2 & mask3, "price"] = wine.price * 0.8

### vieweing the results
wine.loc[mask1 & mask2 & mask3]
```

#### ... and create a new columns called "SalesTitle", adding: "SALE: " in the beginning of each Title for these wines and "NOT ON SALES: " to the remaining ones.

```python
wine.loc[wine["variety"] == "Sangiovese", "SalesTitle"] = "SALE: " + wine.title
wine.loc[wine["variety"] != "Sangiovese", "SalesTitle"] = "NOT ON SALE: " + wine.title

wine.head(100)
```
